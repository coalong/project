```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib
```

```python
data = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv',index_col= 0) 
#데이터 불러오기
```

```python
data 
#데이터 형태 보기 (16598개 데이터와 9개 컬럼)
# Name : 게임의 이름
# Platform : 게임 지원 플랫폼 이름
# Year :  출시 연도
# Genre : 게임 장르
# Publisher : 게임 제작 회사
# NA_Sales : 북미지역 출고량
# EU_Sales : 유럽지역 출고량
# JP_Sales : 일본지역 출고량
# Other_Sales : 기타지역 출고량
```

```python
data.isnull().sum()
#결측치 갯수 확인
```

```python
data=data.dropna()   # 결측치 삭제
data.isnull().sum()  #없어졌나 확인
data=data.drop_duplicates() #완전한 중복 데이터 제거거
```

```python
data.dtypes  #데이터 타입 확인
```

```python
data = data.astype({'Year': 'int'})
# 출시연도 실수형에서 정수형으로 타입 변환
```

```python
#출고량 데이터 단위 맞추기
```

```python
j=[]
for i in data.NA_Sales:
    if 'K' in i:
        i=i.replace("K","")
        j.append(float(i)*0.001)
    elif 'M' in i:
        i=i.replace("M",'')
        j.append(float(i))
    else:
        j.append(float(i))
data['북미 출고량'] = j
```

```python
j=[]
for i in data.EU_Sales:
    if 'K' in i:
        i=i.replace("K","")
        j.append(float(i)*0.001)
    elif 'M' in i:
        i=i.replace("M",'')
        j.append(float(i))
    else:
        j.append(float(i))
data['유럽 출고량'] = j
```

```python
j=[]
for i in data.JP_Sales:
    if 'K' in i:
        i=i.replace("K","")
        j.append(float(i)*0.001)
    elif 'M' in i:
        i=i.replace("M",'')
        j.append(float(i))
    else:
        j.append(float(i))
data['일본 출고량'] = j
```

```python
j=[]
for i in data.Other_Sales:
    if 'K' in i:
        i=i.replace("K","")
        j.append(float(i)*0.001)
    elif 'M' in i:
        i=i.replace("M",'')
        j.append(float(i))
    else:
        j.append(float(i))
data['기타 출고량'] = j
```

```python
data
```

```python
data.drop(columns=["NA_Sales", "EU_Sales","JP_Sales","Other_Sales"],inplace=True)
```

```python
j=[]
for i in range(1,16599):
    try:
        j.append(data['북미 출고량'][i] + data['일본 출고량'][i] + data['유럽 출고량'][i] + data['기타 출고량'][i])
    except:
        pass
```

```python
data['전체 출고량']= j #전체출고량 변수 생성
```
